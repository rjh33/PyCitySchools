```python

#student data by school
groupedStudentBySchooldf = student_data.groupby(['school_name'], as_index=False)

groupedStudentBySchooldf.head()

```

```python
#mean math and reading scores

schoolMeanMathdf=pd.DataFrame(groupedStudentBySchooldf['math_score'].mean())

#schoolMeanMathdf.head()

schoolMeanReadingdf=pd.DataFrame(groupedStudentBySchooldf['reading_score'].mean())

#schoolMeanReadingdf.head()
```

```python
#I can't yet figure this part out.... the percentage above 70
#figure out the number above 70
PassingMath=student_data[student_data['math_score']>=70].groupby("school_name", as_index=False)
mathPassDF=pd.DataFrame(PassingMath["math_score"].count())
mathPassDF.columns=["school_name", "mathCount"]

mathPassDF.head()

passingRead=student_data[student_data['reading_score']>=70].groupby("school_name", as_index=False)
readPassDF=pd.DataFrame(passingRead["reading_score"].count())
readPassDF.columns=["school_name", "readCount"]

readPassDF.head()
```

```python
#merge the data tables school, avg math score, and average reading score

mergeStudentBySchooldf=pd.merge(school_data, schoolMeanMathdf, on="school_name")

mergeStudentBySchooldf=pd.merge(mergeStudentBySchooldf, schoolMeanReadingdf, on= "school_name")
mergeStudentBySchooldf=pd.merge(mergeStudentBySchooldf, mathPassDF, on= "school_name")
mergeStudentBySchooldf=pd.merge(mergeStudentBySchooldf, readPassDF, on= "school_name")
mergeStudentBySchooldf.head()
```

```python
#budget per student
mergeStudentBySchooldf["Budget per Student"]=mergeStudentBySchooldf["budget"]/mergeStudentBySchooldf['size']

mergeStudentBySchooldf.head()
```

```python
#% students pass math and pass reading

mergeStudentBySchooldf["Math Passing Percent"]=mergeStudentBySchooldf['mathCount']/mergeStudentBySchooldf['size']*100
mergeStudentBySchooldf["Reading Passing Percent"]=mergeStudentBySchooldf['readCount']/mergeStudentBySchooldf['size']*100

#overall passing rate
mergeStudentBySchooldf["Overall Pass Rate"]=(mergeStudentBySchooldf["Math Passing Percent"]+mergeStudentBySchooldf["Reading Passing Percent"])/2

mergeStudentBySchooldf.head()
```

```python
schoolFinalSummary=pd.DataFrame(mergeStudentBySchooldf[["school_name", "type", "size", "budget", "Budget per Student", "math_score","reading_score","Math Passing Percent","Reading Passing Percent","Overall Pass Rate"]])
schoolFinalSummary.head()
```
